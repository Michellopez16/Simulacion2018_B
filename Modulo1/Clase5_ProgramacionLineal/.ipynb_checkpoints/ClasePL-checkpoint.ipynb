#  Introducción a Programación lineal

La programación lineal es el campo de la optimización matemática dedicado a maximizar o minimizar (optimizar) una función lineal, denominada función objetivo, de tal forma que las variables de dicha función estén sujetas a una serie de restricciones expresadas mediante un sistema de ecuaciones o inecuaciones también lineales. 

## Problemas de programación lineal expresados en **forma canonica **



https://en.wikipedia.org/wiki/Linear_programming

# Ejemplo gráfico:

## Problema

La compañia de Pepe diseña y vende dos tipos de anillos: **AA y TT**. La empresa puede producir hasta **24 anillos** cada día, usando como máximo un total de **60 h** de mano de obra. Le toma **3h un anillo AA** y **2h uno del tipo TT**, y además el beneficio por cada anillo **AA es $\$40$** y por anillo **TT es $\$30$** <span class="girk">maximice el beneficio</span>. 

|   |Horas|Cantidad|Beneficio|
|---|---|---|---|
|Anillo AA|3h|x|40|
|Anillo TT|2h|y|30|

- Función beneficio(objetivo)

$$f(x,y) = 40x+30y$$

- Restricciones.
$$ x+y\leq 24$$
$$3x+2y\leq 60$$
$$x\geq 0$$
$$y\geq 0$$

## Resolverlo de manera gráfica

- Generamos las lineas
$$ x+y = 24$$
$$3x+2y = 60$$

- Despejando
$$ y_1 = 24 -x$$
$$ y_2 = \frac{60-3x}{2} $$

```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
```

```python
x = np.arange(-1,25)
y1 = 24-x
y2 = (60-3*x)/2
f = lambda X,Y: 40*np.array(X)+30*np.array(Y)
```

```python
fig = plt.figure()
ax = fig.gca()
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['left'].set_smart_bounds(True)
ax.spines['bottom'].set_smart_bounds(True)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.grid(True, zorder=5)

ax.plot(x,y1)
ax.plot(x,y2)
ax.fill_between(x,y1,where=x>=0,facecolor="blue",alpha=0.3,zorder=6 )
ax.fill_between(x,y2,where=x>=0,facecolor="green",alpha=0.3,zorder=7 )
ax.plot([0,12,20],[24,12,0],'ro')

ax.annotate("(0,24)", xy=(0, 24), xytext=(5, 25),
    arrowprops=dict(arrowstyle="->"))

ax.annotate("(12,12)", xy=(12, 12), xytext=(15, 15),
    arrowprops=dict(arrowstyle="->"))
ax.annotate("(20,0)", xy=(20, 0), xytext=(20, 6),
    arrowprops=dict(arrowstyle="->"))


plt.show()
```

### Las ecuaciones las ponemos de manera matricial
$$ x+y = 24$$
$$3x+2y = 60$$

```python
A=[ [1,1],[3,2]]
b=[24,60]
np.linalg.solve(A,b) ## Solucion de nuestro sistema
```

```python
# evaluamos todas las soluciones 
f([0,12,20],[24,12,0])
# La mayor es f(12,12) = 840
```

# Algoritmo Simplex

El término algoritmo Símplex habitualmente se refiere a un conjunto de métodos muy usados para resolver problemas de programación lineal, en los cuales se busca el máximo de una función lineal sobre un conjunto de variables que satisfaga un conjunto de inecuaciones lineales. 



donde **$x$** son las variables desde la forma estándar, **$x_s$** son las variables de holgura introducidas en el proceso de aumentación, **$c$** contiene los coeficientes de optimización, describe el sistema de ecuaciones contraídas, y **$z$** es la variable a ser maximizada.

# Librerías de Python para problemas de optimización

Como es de esperar, en el ecosistema científico de Python podemos encontrar varias librerías que nos van a ayudar a enfrentar los problemas de optimización, entre las que podemos destacar:

Debemos destacar que tanto PuLP como Pyomo requieren la instalación adicional de diferentes solvers para poder resolver los problemas de optimización. Algunos de los solvers que soportan son: GLPK, COIN CLP/CBC, CPLEX y GURUBI, entre otros.

## Utilizando pulp

PuLP is a free open source software written in Python. It is used to describe optimisation problems as mathematical models. PuLP can then call any of numerous external LP solvers (CBC, GLPK, CPLEX, Gurobi etc) to solve this model and then use python commands to manipulate and display the solution.

https://pythonhosted.org/PuLP/main/installing_pulp_at_home.html#installation

```python
!pip install pulp
```

```python
from pulp import *
```

```python
pulp.LpVariable?
```

```python
mi_lp_problema = pulp.LpProblem("Mi LP Problema", pulp.LpMinimize)

x = pulp.LpVariable('x',lowBound=0,cat='Continuous')
y = pulp.LpVariable('y',lowBound=0,cat='Continuous')

# Objective function
mi_lp_problema +=  x +  2*y

# Constraints
mi_lp_problema += 3*x + 4*y >= 1
mi_lp_problema += 2*x + 5*y >= 2
```

```python
mi_lp_problema
```

```python
mi_lp_problema.solve()
pulp.LpStatus[mi_lp_problema.status]
```

```python
for variable in mi_lp_problema.variables():
    print("{} = {}".format(variable.name, variable.varValue))
```

```python
print(pulp.value(mi_lp_problema.objective))
```

```python
from pyomo.environ import *
```

```python
m = ConcreteModel()

m.x_1 = Var(within=NonNegativeReals)
m.x_2 = Var(within=NonNegativeReals)

m.obj = Objective(expr= m.x_1 + 2*m.x_2, sense=minimize )

m.c1 = Constraint(expr= 3*m.x_1 + 4*m.x_2 >= 1)
m.c2 = Constraint(expr= 2*m.x_1 + 5*m.x_2 >= 2)

solver = SolverFactory('ipopt')
status = solver.solve(m)

print("Status = %s" % status.solver.termination_condition)

print("%s = %f" % (m.x_1, value(m.x_1)))
print("%s = %f" % (m.x_2, value(m.x_2)))
print("Objective = %f" % value(m.obj))
```

## Utilizando scipy.optimize 

```python
c = [1, 2]
A = [[-3, -4], 
     [-2, -5]]

b = [-1, -2]

x0_bounds = (0, None)
x1_bounds = (0, None)

from scipy.optimize import linprog

res = linprog(c, A_ub=A, b_ub=b, bounds=(x0_bounds, x1_bounds),
               options={"disp": True})

print(res)
```

## Con Pulp
$$
\begin{array}{ll}
\mbox{minimize}   &  2x_1 + x_2 \\
\mbox{subject to} &   -x_1  + x_2 \leq 1 \\
       & x_1  + x_2 \geq 2 \\
       & x_2 \geq 0 \\
       & x_1 -2x_2 \leq 4
\end{array}
$$

```python
# pulp 
mi_lp_problema = pulp.LpProblem("Mi LP Problema", pulp.LpMinimize)

x = pulp.LpVariable('x',lowBound=0,cat='Continuous')
y = pulp.LpVariable('y',lowBound=0,cat='Continuous')

# Objective function
mi_lp_problema +=  2*x +  y

# Constraints
mi_lp_problema += -x + y <= 1
mi_lp_problema +=  x + y >= 2
mi_lp_problema +=  x - 2*y <= 4

mi_lp_problema.solve()
pulp.LpStatus[mi_lp_problema.status]

for variable in mi_lp_problema.variables():
    print("{} = {}".format(variable.name, variable.varValue))
    
print(pulp.value(mi_lp_problema.objective))
```

## Con Pyomo
$$
\begin{array}{ll}
\mbox{minimize}   &  2x_1 + x_2 \\
\mbox{subject to} &   -x_1  + x_2 \leq 1 \\
       & x_1  + x_2 \geq 2 \\
       & x_2 \geq 0 \\
       & x_1 -2x_2 \leq 4
\end{array}
$$

```python
# pyomo 
m = ConcreteModel()

m.x_1 = Var(within=NonNegativeReals)
m.x_2 = Var(within=NonNegativeReals)

m.obj = Objective(expr= 2*m.x_1 + m.x_2, sense=minimize )

m.c1 = Constraint(expr= -m.x_1 + m.x_2 <= 1)
m.c2 = Constraint(expr= m.x_1 + m.x_2 >= 2)
m.c3 = Constraint(expr= m.x_1 - 2*m.x_2 <= 4)

solver = SolverFactory('ipopt')
status = solver.solve(m)

print("Status = %s" % status.solver.termination_condition)

print("%s = %f" % (m.x_1, value(m.x_1)))
print("%s = %f" % (m.x_2, value(m.x_2)))
print("Objective = %f" % value(m.obj))
```

## Con CVXOPT
$$
\begin{array}{ll}
\mbox{minimize}   &  2x_1 + x_2 \\
\mbox{subject to} &   -x_1  + x_2 \leq 1 \\
       & x_1  + x_2 \geq 2 \\
       & x_2 \geq 0 \\
       & x_1 -2x_2 \leq 4
\end{array}
$$

```python
from cvxopt import matrix, solvers
A = matrix([ [-1.0, -1.0, 0.0, 1.0], 
            [1.0, -1.0, -1.0, -2.0] ])

b = matrix([ 1.0, -2.0, 0.0, 4.0 ])
c = matrix([ 2.0, 1.0 ])
sol=solvers.lp(c,A,b)
```

```python
print("[x1,x2] = ",sol['x'].T)
print ("Objective ",sol['primal objective'])
```

## scipy.optimize 

```python
c = [2, 1]
A = [[-1, 1], 
     [-1, -1],
     [1, -2]]

b = [1, -2, 4]

x0_bounds = (0, None)
x1_bounds = (0, None)

from scipy.optimize import linprog

res = linprog(c, A_ub=A, b_ub=b, bounds=(x0_bounds, x1_bounds),
               options={"disp": True})

print(res)
```
