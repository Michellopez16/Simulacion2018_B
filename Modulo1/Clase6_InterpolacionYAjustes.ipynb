```python
from scipy.stats import linregress
x = np.array([-1,0,1,2])
y = np.array([0,1,2,1])

res = linregress(x, y)

xn = np.arange(-1,2,.01)
yn= res.intercept + res.slope*xn

plt.figure(figsize=(8,6))
plt.plot(x,y,'*')
plt.plot(xn,yn)
plt.plot(x, y, '*b')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```

```python
x = np.linspace(0, 1, 30)
y = 10*x + 2 + np.random.randn(30)


res = linregress(x, y)
xn = np.arange(0,1,.01)
yn= res.intercept + res.slope*xn

plt.figure(figsize=(8,6))
plt.plot(xn,yn)
plt.plot(x, y, '*b')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```

```python
from scipy.optimize import curve_fit

def acetona(x, a, b, c):
    return a+b*x+c*(x**2)

T = np.array([56.5,113.0,181.0,214.5])
P = np.array([1,5,20,40])  

(A, B, C), _ = curve_fit(acetona, P, T)

print(A, B, C)

#nxdat = np.linspace(0, 15, 50)
nxdat = np.arange(0,40,.1)
ydatAjuste = acetona(nxdat, A, B, C)

plt.plot(P, T,"o",  nxdat, ydatAjuste) 
plt.legend(["Datos",  "Ajuste"], loc="best");
```

```python
acetona(4, A, B, C)
```
